```python
import pandas as pd
import csv
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df = pd.read_csv("Order1.csv");
transaction = df.groupby("Member_number")["itemDescription"].apply(list).tolist()
```

```python
te = TransactionEncoder()
x = te.fit_transform(transaction)
df = pd.DataFrame(x,columns=te.columns_)
df
```

```python
af = apriori(df,min_support= 0.01, use_colnames=True)
```

```python
af
```

```python
rules = association_rules(af, metric='confidence', min_threshold=0.50)
rules
```
