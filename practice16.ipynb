```python
import pandas as pd
import csv
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
dataset = []
with open("Order2.csv") as file:
    reader = csv.reader(file);
    for row in reader:
        dataset+=[row]
```

```python
te = TransactionEncoder();
x = te.fit_transform(dataset);
```

```python
help(association_rules)
```

```python
df = pd.DataFrame(x,columns = te.columns_)
```

```python
df.head()
len(df)
```

```python
a = apriori(df, min_support=0.01, use_colnames=True)
a
```

```python
rules = association_rules(a,metric = 'confidence',min_threshold=0.25)
```
