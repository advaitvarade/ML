```python
from sklearn.cluster import KMeans,AgglomerativeClustering
from matplotlib import pyplot as plt
import numpy as np
```

```python
points = np.array( [[0.1,0.6 ] ,[0.1,0.6] ,[0.15,0.71],
[0.08,0.9] ,[0.16, 0.85] ,[0.2,0.3] ,[0.25,0.5]])
```

```python
center = np.array([[0.1,0.6],[0.3,0.2]])
```

```python
kmean = KMeans(n_clusters=2,init=center,n_init=1,random_state=0)
ans=kmean.fit_predict(points)
```

```python
print("P6 belongs to :",ans[6])
print("centories:",kmean.cluster_centers_)
```

```python
plt.scatter(points[:,0],points[:,1],c=ans)
```
