```python
import pandas as pd
import scipy.cluster.hierarchy as sch
```

```python
data = pd.read_csv('hitters.csv')
data.head()
```

```python
data = data.dropna(subset=['CRuns'])
```

```python
x=data[['CRuns']]
```

```python
import matplotlib.pyplot as plt
sch.dendrogram(sch.linkage(x,method="ward"))
```
