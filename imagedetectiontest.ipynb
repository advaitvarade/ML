```python
from ultralytics import YOLO
import cv2
```

```python
model=YOLO("yolov8n.pt")
```

```python
cap=cv2.VideoCapture(0)
```

```python
while True:
    ret, frame =cap.read()
    if not ret:
        break
    results = model(frame)
    annotated_frame = results[0].plot()
    cv2.imshow("YOLO Webcam", annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
```

```python
cap.release()
cv2.destroyAllWindows()
```
