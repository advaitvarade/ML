```python
import pandas as pd 
from sklearn.cluster import KMeans,AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import numpy as ny
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv("Cities_r2.csv")
```

```python
x= df[["total_graduates"]]
se = StandardScaler()
x = se.fit_transform(x);
```

```python
insertia=[]
for i in range(2,11):
    km = KMeans(n_clusters=i,random_state=0, n_init=1)
    km.fit_predict(x)
    insertia.append(km.inertia_)
```

```python
plt.plot((range(2,11)),insertia,marker=".")
plt.grid()
plt.xticks(range(2,11))
```

```python
k=KMeans(n_clusters=2)
df["a"]= k.fit_predict(x)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["total_graduates"],df['population_total'],c=df['a'])
```
