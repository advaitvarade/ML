```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("bank_transactions_data_2.csv")
df.head()
len(df)
```

```python
le = LabelEncoder()
df["TransactionType"] = le.fit_transform(df["TransactionType"])
df["Channel"] = le.fit_transform(df["Channel"])
df['CustomerOccupation']=le.fit_transform(df['CustomerOccupation']);
x = df[["TransactionAmount","TransactionType","Channel","CustomerAge","AccountBalance"]]
y = df['CustomerOccupation']

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.25 , random_state=0)
```

```python
from sklearn.svm import SVC
model = SVC(kernel="poly")
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score
accuracy_score(y_pred,y_test)
```
